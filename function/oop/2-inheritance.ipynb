# Inheritance In Python

It is fundamental concept is Object-Oriented Programming (OOP) that allow a class to inherit attributes and methods from another class.


```python
# single inheritance 

class Car:
    def __init__(self,name,engine_type,is_self_driving):
        self.name = name
        self.engine_type = engine_type
        self.is_self_driving = is_self_driving
        
    def mechanism(self):
        print(f"Car Name : {self.name}")
        print(f"Engine Type : {self.engine_type}")
        
    def drive(self):
        print(f"{self.name} car is driving : {self.is_self_driving}")
        
```

```python
class Tata(Car):
    def __init__( self, name, engine_type, is_self_driving):
        super().__init__(name, engine_type, is_self_driving)
tata1 = Tata("TATA","Petrol",False)
tata1.drive()

class Tesla(Car):
    def __init__(self, name, engine_type, is_self_driving):
        super().__init__(name,engine_type,is_self_driving)
tesla1 = Tesla("Tesla","Electric",True)
tesla1.drive()
```

```python
# multiple inheritance 
class Animal:
     def __init__(self,name):
         self.name = name
     def speak(self):
        print(f"{self.name} is speaking ")
        
class Pet:
    def __init__(self,name):
        self.name = name
    def pet_name(self):
        print(f"pet : {self.name}")

class Dog(Animal,Pet):
    def __init(self,name):
        Animal().__init__(name)
        Pet().__init__(name)
        
```

```python
dog1 = Dog('Lucy')
dog1.speak()
dog1.pet_name()
```

```python
class Persion:
    def __init__(self, name, age):
        self.name = name
        self.age  = age

class Employee(Persion):
    def __init__(self, name, age, emp_id):
        super().__init__(name, age)
        self.emp_id = emp_id
        
```

```python
emp1 = Employee("Clarke",24,25)
print(f"{emp1.name}\n{emp1.age}\n{emp1.emp_id}")
emp2 = Employee("Michel",23,45)
print(f"{emp2.name}\n{emp2.age}\n{emp2.emp_id}")
```
