# Encapsulation

Encapsulation is the concept of wrapping data (variable) and methods (functions) as a single unit. It restricts the direct access to some of the object's components, which is a means of preventing accidental interfaces and misuse of the data.

```python
class Person:
    def __init__(self, name, age):
        self.name = name            # pulbic
        self.__age = age            # private
    def print_person(self):
        print(self.name,self.__age)

        
```

```python
p = Person("srikanta",21)
# p.__age = 23
p.print_person()
```

```python
class Person:
    def __init__(self, name, age):
        self._name = name           # Protected
        self._age = age             # Protected
    def print_per(self):
        print(f"{self._name,self._age}")

class Employee(Person):
    def __init__(self,name,age,emp_id):
        super().__init__(name,age)
        self._emp_id = emp_id
    def print_per(self):
        print(f"{self._name,self._age,self._emp_id}")    
```

```python
p = Person('clarke',24)
print(p._name)              # technically accessible, but not recommended
p.print_per()          
e = Employee('cl',24,677)
e.print_per()
```

#### getter & setter method

```python
class Person:
    def setPerson(self,name,age):
        self.__name = name
        self.__age = age
    def getPerson(self):
        return f"{self.__name} {self.__age}"
```

```python
p1 = Person()
p1.setPerson("Mahes",24)
p1.getPerson()
```

```python
class Employee:
    def __init__(self, name, age):
        self.name = name
        self.age = age

    def set_name(self,name):
        self.__name = name
    def set_age(self,age):
        self.__age = age
    def set_emp_id(self,id):
        self.__emp_id = id
    def get_name(self):
        return self.__name
    def get_age(self):
        return self.__age
    def get_emp_id(self):
        return self.__emp_id
```

```python
e = Employee()
e.set_name("Name")
e.set_age(4)
e.set_emp_id(0)
print(e.get_name(),e.get_age(),e.get_emp_id())
```
